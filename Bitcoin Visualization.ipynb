```python
import pandas_datareader as web
import matplotlib.pyplot as plt
import mplfinance as mpf
import datetime as dt
```

```python
crypto = "BTC"
currency = "USD"
```

```python
start = dt.datetime(2020,1,1)
end = dt.datetime.now()
data = web.DataReader(f"{crypto}-{currency}","yahoo")
print(data)
```

```python
plt.plot(data['Close'])
# plt.rcParams['figure.figsize'] = (20,36)
plt.show()
```

```python
mpf.plot(data, type = 'candlestick')
```

```python
btc = web.DataReader(f"{crypto}-{currency}","yahoo")
eth = web.DataReader(f"ETH-{currency}","yahoo")
plt.plot(btc['Close'], label = "BTC")
plt.plot(eth['Close'], label = "ETH")
plt.yscale('log')
plt.legend(loc = 'upper left')
plt.show()
```
